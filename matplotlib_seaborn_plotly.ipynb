#Matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import plotly
import plotly.express as px

import seaborn as sns
sns.set()
```

##Wykres linowy

```python
fig = plt.figure()
fig.show()
```

```python
plt.plot([0,1], [0,3])
```

```python
plt.plot([0,1], [0,3], label='line', color='green')
plt.legend()
```

```python
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
plt.plot([0,1, 2], [0,3, 2], label='line', color='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres')
plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

##Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 10])
```

```python
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 10], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [4, 5, 6], 'ro')
```

#Plotly Express

##Iris data

```python
df = px.data.iris()
df.head()
```

```python
px.scatter(data_frame=df, x='sepal_length', y='sepal_width')
```

```python
px.scatter(data_frame=df, x='sepal_length', y='sepal_width', width=500, height=400)
```

```python
px.scatter(data_frame=df, x='sepal_length', y='sepal_width', width=500, height=400, color='species')
```

```python
px.scatter(data_frame=df, x='sepal_length', y='sepal_width', width=500, 
           height=400, color='species', marginal_x='violin')
```

```python
px.scatter(data_frame=df, x='sepal_length', y='sepal_width', width=1000, 
           height=700, color='species', marginal_x='violin', marginal_y='box',
           title='Iris', trendline='ols')
```

```python
px.scatter_matrix(data_frame=df, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                  color='species', title='Iris')
```

```python
px.parallel_coordinates(data_frame=df, color='species_id')
```

##Tips data

```python
px.scatter(data_frame=df, x='total_bill', y='tip', facet_row='time',
           trendline='ols', color='smoker')
```

##Gapminder data

```python
df = px.data.gapminder()
df.info()
```

```python
px.scatter(df.query('year == 2007'), x='gdpPercap', y='lifeExp', 
           size='pop', color='continent', log_x=True, 
           hover_name='country', size_max=60)
```

```python
px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', 
           log_x=True, hover_name='country', size_max=60, animation_frame='year',
           range_y=[25, 90])
```

```python
px.line(data_frame=df.query("country == 'Poland'"), x='year', y='pop')
```

```python
px.line(data_frame=df.query("continent == 'Europe'"), x='year', y='pop', color='country')
```
